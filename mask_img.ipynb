```python
import os
import cv2 as cv
import numpy as np
import matplotlib.cm as cm
src_dir = './pure_testresult/'

for root, dirs, files in os.walk(src_dir):
    for f in files:
        if f.startswith('src'):
            src = cv.imread(root + f)
            name = f[3:]
            mask = cv.imread(root + 'pred' + name)
            mask[:,:,0] = 0
            mask[:,:,1] = 0
            mask = mask * 255
            result = 0.4*mask + 0.6 * src
            print name+'processing'
            #result = cv.addWeighted(src, 1, mask, 0.4, 0)
            cv.imwrite(src_dir+'mask'+name, result)
                
print 'done'
        
                           
```
