```python
import cv2 as cv
import os

src_dir = './test/images/'
tar_dir = './test/images_clip/'
IMAGE_SIZE = 500

for root, dirs, files in os.walk(src_dir):
    for f in files:
        if f.endswith('jpg'):
            src = cv.imread(root + f)
            name = f[:-4]
            print name+'processing'
            for i in range(5000/IMAGE_SIZE):
                for j in range(5000/IMAGE_SIZE):
                    cv.imwrite(tar_dir+name+'_'+str(i)+'_'+str(j)+'.jpg',src[i*IMAGE_SIZE:(i+1)*IMAGE_SIZE,j*IMAGE_SIZE:(j+1)*IMAGE_SIZE])
                
print 'done'
        
                           
```
